## Compiling the experiment

To compile an experiment with [`ettk`](https://github.com/NLP-HLP/ettk), we need an experiment directory with a configuration file and any required materials and resources. Have a look at the `pronoun-experiment` directory next to this notebook and the `config.yaml` file inside it:

```python
!cat pronoun-experiment/config.yaml
```

Compiling the experiment generates several files:
- images and areas of interest (AOIs) for each screen (`pronoun-experiment/stimuli/*`)
- a JSON file for each participant defining the order of the screens (`pronoun-experiment/procedures/*`)
- a JSON file containing metadata about the experiment (`pronoun-experiment/experiment.json`)

```python
!python -m ettk.compiler pronoun-experiment
```

If you want to visualize the AOIs, you can add `--area-images` to the command. This will generate additional images in `pronoun-experiment/stimuli` for each type of AOI (e.g., character-level and word-level):

```python
!python -m ettk.compiler pronoun-experiment --area-images
```
